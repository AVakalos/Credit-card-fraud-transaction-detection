```python
X = df.drop(['Class'],axis=1)
y = df['Class']
```

```python
def evaluate(clf):
    
    name = models[clf]
    clf.fit(train,labels)
    pred = clf.predict(X_test) 
    
    precision_list.append(precision_score(y_test, pred))
    recall_list.append(recall_score(y_test,pred))
    
    print('-'*55)
    print('\n' + name + ' Classifier\n')
    print('-'*55)
    
    #print('Precision: {}'.format(precision))
    #print('Recall: {}'.format(recall))
    
    print(classification_report(y_test,pred))
    
    cf = confusion_matrix(y_test, pred)
    print('True Positives: {}'.format(cf[1][1]))
    print('False Negatives: {}'.format(cf[1][0]))
    print('True Negatives: {}'.format(cf[0][0]))
    print('False Positives: {}'.format(cf[0][1]))
    
    fig, ax = plt.subplots(figsize=(7,5))

    plot_precision_recall_curve(clf,X_test,y_test,ax=ax)
    ax.set_title(name + ' Classifier Precision-Recall curve', fontsize=12)
    plt.show()
    print('\n\n')
   
```

```python
models = {}
precision_list = []
recall_list = []

train = X_train_u
labels = y_train_u

lr = LogisticRegression(max_iter=1000)
models[lr] = 'Logistic Regression'

svm = SVC()
models[svm] = 'Support Vector'

knn = KNeighborsClassifier(n_neighbors=3,n_jobs=-1)
models[knn] = 'kNN'

xgb = XGBClassifier(use_label_encoder=False)
models[xgb] = 'XGBoost'

mlp = MLPClassifier(hidden_layer_sizes=(200,200),max_iter = 10000)
models[mlp] = 'MLP'
```

```python
for model in models:
    evaluate(model)
```

```python
fig, ax = plt.subplots(1,2,figsize=(20,8))

sns.barplot(list(models.values()),precision_list,ax=ax[0])
ax[0].set_title('Precision of Classifiers')
  

sns.barplot(list(models.values()),recall_list,ax=ax[1])
ax[1].set_title('Recall of Classifiers')

plt.show()
```

```python
from sklearn.model_selection import StratifiedKFold

kf = StratifiedKFold(n_splits=5,shuffle=True)#,random_state=20)

resampler_kf = SMOTE()


clf_kf = XGBClassifier(use_label_encoder=False)
#clf_kf = KNeighborsClassifier(n_neighbors=3,n_jobs=-1)

split_counter = 1
precisions_list = []
recalls_list = []

for train_index, test_index in kf.split(X,y):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]
    
    #X_train, y_train = resampler_kf.fit_resample(X_train,y_train)
    
    clf_kf.fit(X_train,y_train)
    fold_pred = clf_kf.predict(X_test)
    fold_precision = precision_score(y_test, fold_pred)
    fold_recall = recall_score(y_test, fold_pred)
    precisions_list.append(fold_precision)
    recalls_list.append(fold_recall)
    
    print('\nFold {} Precision: {}  Recall: {}\n'.format(split_counter, fold_precision, fold_recall))
    split_counter = split_counter + 1

print('\n'+55*'-')   
print('Total performance for {}-Fold Cross Validation'.format(kf.get_n_splits()))
print('\nAverage precision: {}  Average Recall: {}\n'.format(np.mean(precisions_list),np.mean(recalls_list)))
```
